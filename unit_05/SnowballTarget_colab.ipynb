```python
# Step 5: Train the agent
!mlagents-learn ./config/ppo/SnowballTarget.yaml \
    --env=./training-envs-executables/linux/SnowballTarget/SnowballTarget \
    --run-id="SnowballTarget" \
    --no-graphics
```

```python
import os

# Ensure the directory exists
os.makedirs('/content/ml-agents/config/ppo/', exist_ok=True)

# Create the SnowballTarget.yaml configuration
config_content = """
behaviors:
  SnowballTarget:
    trainer_type: ppo
    hyperparameters:
      batch_size: 128
      buffer_size: 2048
      learning_rate: 0.0003
      beta: 0.005
      epsilon: 0.2
      lambd: 0.95
      num_epoch: 3
      learning_rate_schedule: linear
    network_settings:
      normalize: false
      hidden_units: 128
      num_layers: 2
      vis_encode_type: simple
    reward_signals:
      extrinsic:
        gamma: 0.99
        strength: 1.0
    keep_checkpoints: 5
    max_steps: 500000
    time_horizon: 64
    summary_freq: 10000
"""

with open('/content/ml-agents/config/ppo/SnowballTarget.yaml', 'w') as f:
    f.write(config_content.strip())

print("Config file created at /content/ml-agents/config/ppo/SnowballTarget.yaml")
```
